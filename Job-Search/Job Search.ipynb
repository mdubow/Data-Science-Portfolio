```python
import pandas as pd
import numpy as np

# here are links for following files:
# https://www.kaggle.com/andrewmvd/data-scientist-jobs
# https://www.kaggle.com/andrewmvd/data-analyst-jobs
# https://www.kaggle.com/andrewmvd/data-engineer-jobs
```

```python
df_analyst = pd.read_csv('input/DataAnalyst.csv', index_col=0)
df_engineer = pd.read_csv('input/DataEngineer.csv', index_col=0)
df_scientist = pd.read_csv('input/DataScientist.csv', index_col=0)

df = pd.concat([df_analyst, df_engineer, df_scientist])
df.head()
```

```python
df.describe(include='all')
```

```python
#drop unnecessary columns
drop_cols = ['Rating', 'Headquarters', 'Founded', 'Type of ownership', 'Revenue', 'Competitors', 'Easy Apply'] #'index'
df = df.drop(drop_cols, axis=1)
df.head()
```

```python
# replace -1 with NaN
df = df.replace(-1, np.nan)
df = df.replace('-1', np.nan, regex=True)
df.head()
```

```python
# remove jobs without salary estimates, then reset index
df = df.dropna()
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
# convert salary estimates into usable integers in two columns
hours_per_week = 40
weeks_per_year = 52

for i in range(df.shape[0]):
    salary_estimate = str(df.loc[i,"Salary Estimate"])
    salary_estimate = salary_estimate.replace("$", "")

    if "Per Hour" in salary_estimate:
        lower, upper = salary_estimate.split("-")
        upper, _ = upper.split("Per")
        upper= upper.strip()
        lower = int(lower)*hours_per_week*weeks_per_year*(1/1000)
        upper = int(upper)*hours_per_week*weeks_per_year*(1/1000)

    else:
        lower, upper = salary_estimate.split("-")
        lower = lower.replace("K", "")
        upper, _= upper.split("(")
        upper=upper.replace("K", "")
        upper = upper.strip()

    lower = int(lower)
    upper = int(upper)
    df.loc[i,"salary_estimate_lower_bound"] = lower
    df.loc[i,"salary_estimate_upper_bound"] = upper

df = df.drop('Salary Estimate', axis=1)
df.head()
```

```python
#do same thing with size
for i in range(df.shape[0]):
    size = str(df.loc[i, 'Size'])

    if "to" in size:
        lower, upper = size.split("to")
        lower = lower.strip()
        upper, _ = upper.split('employees')
        upper = upper.strip()

    else:
        lower = 10000
        upper = 10000
    
    df.loc[i,"size_lower_bound"] = lower
    df.loc[i,"size_upper_bound"] = upper

df = df.drop('Size', axis=1)
df.head()
```

```python
df.describe(include='all')
```

```python
# clean up company names and descriptions
for i in range(df.shape[0]):
    name = df.loc[i,"Company Name"]
    if "\n" in str(name):
        name,_ = name.split("\n")
    df.loc[i,"Company Name"] = name
    
df["Job Description"] = df["Job Description"].replace("\n", " ", regex=True)
```

```python
# create function that will strip white spaces and make everything lowercase
def lower_strip(col):
    return str.lower(col.replace(' ', ''))
```

```python
# create copy of dataframe for cleaned columns
df_clean = df.copy()

for col in df.columns:
    df_clean[col] = df_clean[col].astype(str)
    df_clean[col] = df_clean[col].apply(lower_strip)

df_clean.head()
```

```python
# grab indices of titles
indices = pd.Series(df.index, index=df['Job Title'])
indices[:10]
```

```python
# convert all info into long text
df['text'] = ''
for col in df.columns:
    df[col] = df[col].astype(str)
    df['text'] += df[col] + ' '
```

```python
import gensim, re

# TOKENIZER
def tokenizer(input_string):
    return re.findall(r"[\w']+", input_string)

def get_words(idx, df=df):
    words = df.loc[idx, 'tokens']
    return words

df.loc[:, 'tokens'] = df.text.apply(tokenizer)
documents = []
#df['index'] = df['index'].astype(float).astype(int)
#indices = list(df['index'])
# sentences = []
# for item_no, line in enumerate(df['tokens'].values.tolist()):
#     sentences.append(LabeledSentence(line,[item_no]))

tagged = gensim.models.doc2vec.TaggedDocument
documents = [tagged(words=get_words(idx), tags=[idx]) for idx in indices]


# MODEL PARAMETERS
dm = 1 # 1 for distributed memory(default); 0 for dbow 
size = 300
context_window = 50
seed = 42
min_count = 1
alpha = 0.5
max_iter = 200

# BUILD MODEL
model = gensim.models.doc2vec.Doc2Vec(documents = documents,
dm = dm,
alpha = alpha, # initial learning rate
seed = seed,
min_count = min_count, # ignore words with freq less than min_count
max_vocab_size = None,
window = context_window # the number of words before and after to be used as context
)
```

```python
# demonstrate how query docs can be matched to existing corpus

tokens = "a new sentence to match".split()

new_vector = model.infer_vector(tokens)
sims = model.dv.most_similar([new_vector])
sims
```

```python
def get_tokens():
    try:
        file = input("Please enter query file extension")
        doc = docx2python(file)
    except:
        print("Invalid extension; default file chosen")
        file = file_def
        doc = docx2python(file)
    
    df_doc = pd.DataFrame(pd.Series(doc.text))
    tokenized_doc = df_doc[0].apply(tokenizer)
    tokens = tokenized_doc[0]
    return tokens

def get_salary():
    try:
        lower_sal = input("What is your desired annual salary? Please enter minimum")
        upper_sal = input("What is your desired annual salary? Please enter maximum")
    except:
        print("Invalid salary; defaults of 100k and 150k chosen")
        lower_sal = 100000
        upper_sal = 150000
    lower_sal = str(int(lower_sal)/1000)
    upper_sal = str(int(upper_sal)/1000)
    return lower_sal, upper_sal
    
def get_size():
    try:
        lower_size = input("What is your desired company size? Please enter minimum")
        upper_size = input("What is your desired company size? Please enter maximum")
    except:
        print("Invalid size; defaults of 50 and 200 chosen")
        lower_size = 50
        upper_size = 200
    lower_size = str(lower_size)
    upper_size = str(upper_size)
    return lower_size, upper_size
   
```

```python
def get_info():
    tokens = get_tokens()
    lower_sal, upper_sal = get_salary()
    lower_size, upper_size = get_size()
    tokens.append(lower_sal)
    tokens.append(upper_sal)    
    tokens.append(lower_size)    
    tokens.append(upper_size)
    return tokens
```

```python
def match_postings():
    tokens = get_info()
    new_vector = model.infer_vector(tokens)
    sims = model.dv.most_similar([new_vector])
    print(u'SIMILAR/DISSIMILAR DOCS PER MODEL %s:\n' % model)
    for label, index in [('MOST', 0), ('SECOND-MOST', 1), ('MEDIAN', len(sims)//2), ('LEAST', len(sims) - 1)]:
        print(u'%s %s: «%s»\n' % (label, sims[index], ' '.join(df.loc[sims[index][0]].text)))
```
