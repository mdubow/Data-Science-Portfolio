### Bar Crawl

Overview: Using phone accelerometer data, a machine learning model can predict whether an individual is drunk based on the variance in their movement.

Client: Public health organizations, sociologists, universities, younger people. Potential to create app that warns users of heavy alcohol use. 

Data: UCI Detecting Heavy Drinking Dataset https://archive.ics.uci.edu/ml/datasets/Bar+Crawl%3A+Detecting+Heavy+Drinking

Data Description: Accelerometer data for 13 participants involved in a “bar crawl” event. CSV file includes position in 3 axes and time of measurement in milliseconds.  Additionally, separate CSV files for each participant of TAC (transdermal alcohol content) readings, taken roughly every half hour. “Clean” readings are shifted back in time by 45 minutes to account for time it takes to release alcohol through the skin.

Features: variance in 3 axes (continuous numerical), participant ID (categorical string), phone type (categorical string)
Target: "drunk" label (binary categorical)

### Data Wrangling

1) The first two rows of the accelerometer data were dropped, as they contained all zero values and therefore appear to be a data entry error.

2) Time measurements for accelerometer data are taken in milliseconds since UNIX Epoch (01/01/1970). Values converted to correct datetime objects using the pandas .to_datetime function. 

3) TAC readings were concatenated into a single dataframe by looping over the unique participant ID’s. Here, time measurements are changed to datetime objects as well.

4) TAC readings for two participants are found to be identical, indicating a data entry error. One participant is dropped.

5) Variance is calculated from accelerometer data.
 a) Loop over readings for an individual participant.
 b) Take average of TAC readings within a thirty minute window. Because readings happen roughly every half hour, usually only one or two readings are aggregated.
 c) For every one TAC window, there are 3 sub-windows, each representing the variance in movement over a ten minute period.
 d) If the TAC reading is greater than or equal to 0.08 (legal limit), label row as “drunk” (1 in “drunk” column).
 e) Loop over all participants.


6) Resulting dataframe should have observations with datetime, variance in three axes, TAC reading, and “drunk” label.

```python
def triple_variance_calculator(df_tacs, df_acc, pid, delta1=timedelta(minutes=30), delta2=timedelta(minutes=10)):
    df1 = df_tacs[df_tacs.pid == pid].copy()
    df2 = df_acc[df_acc.pid == pid].copy()
    
    start = max(df1['datetime'].min() + delta1, df2['datetime'].min()) + delta1
    end = df1['datetime'].max()
    tac_window = start
    
    X = []
    t = []
    y = []
    
    while tac_window <= end:
    
        accel_window = tac_window
        mask1 = (df1['datetime'] >= tac_window - delta1) & (df1['datetime'] < tac_window)
        mask2 = (df2['datetime'] >= accel_window) & (df2['datetime'] < accel_window + delta2)
        
        for i in range(3):
            mask1 = (df1['datetime'] >= tac_window - delta1) & (df1['datetime'] < tac_window)
            mask2 = (df2['datetime'] >= accel_window) & (df2['datetime'] < accel_window + delta2)
            X.append(df2[mask2][['x', 'y', 'z']].std())
            t.append(accel_window)
            accel_window += delta2
        
#         X.append(df2[mask2][['x', 'y', 'z']].std())
#         t.append(accel_window)
#         accel_window += delta2
        
#         X.append(df2[mask2][['x', 'y', 'z']].std())
#         t.append(accel_window)
#         accel_window += delta2
        
#         X.append(df2[mask2][['x', 'y', 'z']].std())
#         t.append(accel_window)
#         accel_window += delta2
        
        y.append(df1[mask1]['TAC_Reading'].mean())
        tac_window += delta1
        
    df_x = pd.DataFrame(X, index=t)
    df_x = df_x.rename(columns={'x':'x_variance', 'y':'y_variance', 'z':'z_variance'})
    df_x["datetime"] = t
    df_x['TAC_Reading'] = np.repeat(y, 3)
    df_x = df_x.dropna()
    df_x['pid'] = pid
    df_x['drunk'] = np.where(df_x['TAC_Reading'] >= 0.08, 1, 0)
    df_x['var'] = np.sqrt(df_x['x_variance']**2 + df_x['y_variance']**2 + df_x['z_variance']**2)
    return df_x
```

```python
# try to reexamine positional graphs - maybe they represent velocities or accelerations, not position
def triple_velocity_calculator(df_tacs, df_acc, pid, delta1=timedelta(minutes=30), delta2=timedelta(minutes=10)):
    df1 = df_tacs[df_tacs.pid == pid].copy()
    df2 = df_acc[df_acc.pid == pid].copy()
    
    start = max(df1['datetime'].min() + delta1, df2['datetime'].min()) + delta1
    end = df1['datetime'].max()
    tac_window = start
    
    X = []
    t = []
    y = []
    
    x_vel = []
    y_vel = []
    z_vel = []
    
    
    while tac_window <= end:
    
        accel_window = tac_window
        mask1 = (df1['datetime'] >= tac_window - delta1) & (df1['datetime'] < tac_window)
        mask2 = (df2['datetime'] >= accel_window) & (df2['datetime'] < accel_window + delta2)
        
        for i in range(3):
            # integrate over acceleration to get velocity
            #mask2 = (df2['datetime'] >= accel_window) & (df2['datetime'] < accel_window + delta2)
            x_vel.append(float(np.cumsum(df2[mask2]['x'].mean())))
            y_vel.append(float(np.cumsum(df2[mask2]['y'].mean())))
            z_vel.append(float(np.cumsum(df2[mask2]['z'].mean())))
            t.append(accel_window)
            accel_window += delta2
        
        y.append(df1[mask1]['TAC_Reading'].mean())
        tac_window += delta1
    
    X = {'x_velocity' : list(x_vel), 'y_velocity': list(y_vel), 'z_velocity': z_vel}
    df_x = pd.DataFrame(X, index=t)
    #df_x = df_x.rename(columns={'x':'x_velocity', 'y':'y_velocity', 'z':'z_velocity'})
    df_x["datetime"] = t
    df_x['TAC_Reading'] = np.repeat(y, 3)
    df_x = df_x.dropna()
    df_x['pid'] = pid
    df_x['drunk'] = np.where(df_x['TAC_Reading'] >= 0.08, 1, 0)
    #df_x['velocity'] = np.sqrt(df_x['x_velocity']**2 + df_x['y_velocity']**2 + df_x['z_velocity']**2)
    return df_x
```

### Inferential Statistics

1) Z-tests determine there is a statistically significant difference between drunk and sober movement.

```python
# use z-test to test difference between sober and drunk
df_triple_sa = df[df.pid == 'SA0297']
df_triple_sa.head()
```

```python
x_var_drunk = df_triple_sa[df_triple_sa.drunk == 1].x_variance
x_var_sober = df_triple_sa[df_triple_sa.drunk == 0].x_variance

print(len(x_var_drunk))
print(len(x_var_sober))
```

```python
plt.hist(x_var_sober)
plt.xlabel('Sober x-variance')
plt.ylabel('Frequency')
plt.title('Histogram of sober variance')
plt.show()
```

```python
# take random samples from distributions
size = 50
x_var_drunk_sample = np.random.choice(x_var_drunk, size=size)
x_var_sober_sample = np.random.choice(x_var_sober, size=size)

# use Bessel's correction for std
x_var_drunk_std = np.std(x_var_drunk, ddof=1)
x_var_sober_std = np.std(x_var_sober, ddof=1)

x_var_drunk_sample_mean = x_var_drunk_sample.mean()
x_var_sober_sample_mean = x_var_sober_sample.mean()
```

```python
# compute z-score using above variables
# Null hypothesis: no difference between drunk and sober
# Alternate hypothesis: there is a difference
x_var_z = (x_var_drunk_sample_mean - x_var_sober_sample_mean) / \
np.sqrt((x_var_drunk_std**2 / size) + (x_var_sober_std**2 / size))
print('Z-score for x_variance: {}'.format(x_var_z))
```

```python
# use z-score to compute p-value
# scipy.stats.norm has a cdf function which can be used her
# remember this is a two-tailed test, so multiple by two
from scipy.stats import norm

x_var_p = norm.pdf(x_var_z)*2
print('P-value for x_variance: {}'.format(x_var_p))
```

```python
# create cdf plot for x_variance
x_var_sample_diff = x_var_drunk_sample - x_var_sober_sample
x_var_observed_diff = x_var_drunk.mean() - x_var_sober.mean()
plt.hist(x_var_sample_diff, density=True, cumulative=True, edgecolor='black', bins=size)
plt.ylabel('CDF')
plt.xlabel('Sample difference in x-variance')
print('Sample difference in means of x-variance: {}'.format(x_var_drunk_sample_mean - x_var_sober_sample_mean))
print('Observed difference in means of x-variance: {}'.format(x_var_observed_diff))
```

```python
# now pdf for x_variance
plt.hist(x_var_sample_diff, density=True, edgecolor='black')
plt.xlabel('Sample difference in x-variance')
plt.ylabel('PDF')
plt.axvline(x_var_sample_diff.mean(), color='r', label='Sample mean')
plt.axvline(x_var_observed_diff, color='r', linestyle='--', label='Observed mean')
#plt.axvline(-x_var_observed_diff, color='r', linestyle='--', label='Observed mean')
plt.legend()
plt.show()
```

```python
# create cdf plot for z_variance
z_var_sample_diff = z_var_drunk_sample - z_var_sober_sample
z_var_observed_diff = z_var_drunk.mean() - z_var_sober.mean()
plt.hist(z_var_sample_diff, density=True, cumulative=True, edgecolor='black', bins=size)
plt.ylabel('CDF')
plt.xlabel('Sample difference in y-variance')
print('Sample difference in means of y-variance: {}'.format(z_var_drunk_sample_mean - z_var_sober_sample_mean))
print('Observed difference in means of y-variance: {}'.format(z_var_observed_diff))
```

```python
# now pdf for z_variance
plt.hist(z_var_sample_diff, density=True, edgecolor='black')
plt.xlabel('Sample difference in z-variance')
plt.ylabel('PDF')
plt.axvline(z_var_sample_diff.mean(), color='r', label='Sample mean')
plt.axvline(z_var_observed_diff, color='r', linestyle='--', label='Observed mean')
plt.legend()
plt.show()
```

```python
# take random samples from distributions
size = 50
x_var_droid_sample = np.random.choice(x_var_droid, size=size)
x_var_iphone_sample = np.random.choice(x_var_iphone, size=size)

# use Bessel's correction for std
x_var_droid_std = np.std(x_var_droid, ddof=1)
x_var_iphone_std = np.std(x_var_iphone, ddof=1)

x_var_droid_sample_mean = x_var_droid_sample.mean()
x_var_iphone_sample_mean = x_var_iphone_sample.mean()
x_var_phone_sample_diff_mean = x_var_droid_sample_mean - x_var_iphone_sample_mean

# compute z-score using above variables
# Null hypothesis: no difference between phones
# Alternative hypothesis: there is a difference
x_var_phone_z = (x_var_phone_sample_diff_mean) / \
np.sqrt((x_var_droid_std**2 / size) + (x_var_iphone_std**2 / size))
print('Z-score for z_variance: {}'.format(x_var_phone_z))

x_var_phone_p = norm.pdf(x_var_phone_z)*2
print('P-value for z_variance: {}'.format(x_var_phone_p))
```
